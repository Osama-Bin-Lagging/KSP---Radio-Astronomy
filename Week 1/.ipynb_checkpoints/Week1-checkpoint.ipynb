# 3- Plotting The Jet Afterglow Lightcurve Of GW170817

```python
gw170817 = ascii.read('http://www.tauceti.caltech.edu/kunal/gw170817/gw170817_afterglow_data_full.txt')
data = gw170817.to_pandas()
data
```

### Extracting the relevant data
We get the data from VLA telescope with frequency of 3GHz and the data of Chandra telescope with frequency of 2.41 $ \times 10^{17}$Hz

```python
mask = np.where(data['Telescope'] == 'VLA')[0]
mask1 = np.where(data['Freq'] == 3.000000e+09)[0]
mask2 = np.intersect1d(mask, mask1)

Chandra_mask = np.where(data['Telescope'] == 'Chandra')[0]
Chandra_mask1 = np.where(data['Freq'] == 2.410000e+17)[0]
Chandra_mask2 = np.intersect1d(Chandra_mask, Chandra_mask1)
```

#### m = -2.5 log(Flux) + C
let C = 0 (for simplicity VLA, taking C = 0 for Chandra too, which we'll adjust later)

Removing the '<' from the data, and converting all the string data into floats for further processing, and converting the 2.41 $ \times 10^{17}$Hz data to 3 GHz data, with the relation  $F_v \propto v^{-0.584}$

```python
flux_data = np.array(data['FluxD'][mask2])
for index in range(len(flux_data)):
    if flux_data[index][0] == '<':
        flux_data[index] = flux_data[index][1:]
    flux_data[index] = float(flux_data[index])
    
apparent_magnitude = -2.5 * np.log(list(flux_data))

error_VLA = np.abs(-2.5 * np.log(np.array(data['FluxDErr'][mask2])))
error_Chandra = np.abs(-2.5 * np.log(np.array(data['FluxDErr'][Chandra_mask2])))
Chandra_flux= np.array(data['FluxD'][Chandra_mask2])

for index in range(len(Chandra_flux)):
    if Chandra_flux[index][0] == '<':
        Chandra_flux[index] = Chandra_flux[index][1:]
    Chandra_flux[index] = float(Chandra_flux[index])
new_flux = np.array(Chandra_flux)*pow(2.410000e+17/3.000000e+09,-0.584)


Capparent_magnitude = -2.5 * np.log(list(new_flux))
```

### Plotting the scatter plots for the data points

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2, 1)
plt.errorbar(data['T'][mask2], apparent_magnitude, yerr=error_VLA, fmt='<', label='Data')
plt.xlabel('Time (days)')
plt.ylabel('Flux (units)')
plt.title('Lightcurve - VLA')
plt.grid(True)

plt.subplot(1,2, 2)
plt.errorbar(data['T'][Chandra_mask2], Capparent_magnitude, yerr = error_Chandra, fmt='<', label='Data')
plt.xlabel('Time (days)')
plt.ylabel('Flux (units)')
plt.title('Lightcurve - Chandra')
plt.grid(True)

plt.show()
```

### Defining the model function

```python
def linear_plot(x, a, b):
    return a * x + b 
```

```python
params, covar = curve_fit(linear_plot, data['T'][mask2], apparent_magnitude)
a_fitted, b_fitted = params
print('value of alpha for VLA, after linear regression is',a_fitted,'and value of beta for VLA, after linear regression is',b_fitted)

params, covar = curve_fit(linear_plot, data['T'][Chandra_mask2], Capparent_magnitude)
C_a_fitted, C_b_fitted = params
print('value of alpha for Chandra, after linear regression is',C_a_fitted,'and value of beta for Chandra, after linear regression is',C_b_fitted)
```

```python
plt.errorbar(data['T'][mask2], apparent_magnitude, yerr=error_VLA, fmt='bo',label='Data')

diff_margin = sum(Capparent_magnitude)/len(Capparent_magnitude) - sum(apparent_magnitude)/len(apparent_magnitude)

plt.errorbar(data['T'][Chandra_mask2], Capparent_magnitude - diff_margin, yerr = error_Chandra,color ='orange', fmt='<', label='Data')

plt.plot(data['T'][mask2], linear_plot(data['T'][mask2], a_fitted, b_fitted),'b-')
plt.plot(data['T'][Chandra_mask2], linear_plot(data['T'][Chandra_mask2], C_a_fitted, C_b_fitted) - diff_margin, color ='orange')
plt.show()
```
